# Analisis Activos Ponderados por Riesgo de Crédito (IRS) Bancos Chilenos

Instrucciones:

    1. Primero debe ejecutarse el recuadro que extrae los datos del indice IRS para los bancos chilenes, 
    categorizados sus codigos en archivo de texto Codigo_Bancos.txt (archivo de configuración), y al momento de ejecutar
    se le solicita al usuario que ingrese una cantidad de periodos que quiere bajar historicos, partiendo del último dato
    conocido, este numero puede ser minimo 3 y máximo 36 para el análisis.
    
    2. En el segundo paso, se solicita al usuario que ingrese un número del 1 al 4, que representan las cuatro
    posibilidades de cálculo implementadas. El detalle de cada una, aparece al momento de ingresar el número por pantalla.

```python
# 1. Recuadro para extraer datos desde API SBIF, y guardar en archivo de texto local

def carga_datos_sbif(n_periodos):
    
    import requests

    periodos = str(n_periodos-1)
    
    apikey = "c540b05bfed3f6097e5e02c7bfb7ce9b118ab432"
    formato = "json"
    
    bancos = open("Codigo_Bancos.txt", "r")
    datos_indices = open("Indices_IRS.txt", "w")
    
    for linea in bancos:
        institucion = linea.strip()
        consulta_api = requests.get("https://api.sbif.cl/api-sbifv3/recursos_api/adecuacion/regresionmensual/"+periodos+"/instituciones/"+institucion+"/indicadores/irs?apikey="+apikey+"&formato="+formato)
        resultado = consulta_api.json()
        i=0
        while i < len(resultado["AdecuacionDeCapital"]):
            datos_indices.write(resultado["AdecuacionDeCapital"][i]["Institucion"]["CodigoInstitucion"])
            datos_indices.write(";")
            datos_indices.write(resultado["AdecuacionDeCapital"][i]["Institucion"]["NombreInstitucion"])
            datos_indices.write(";")
            datos_indices.write(str(resultado["AdecuacionDeCapital"][i]["Periodo"]["anho"]))
            datos_indices.write(";")
            datos_indices.write(str(resultado["AdecuacionDeCapital"][i]["Periodo"]["mes"]))
            datos_indices.write(";")
            datos_indices.write(resultado["AdecuacionDeCapital"][i]["Indicadores"]["IRS"])
            datos_indices.write("\n")
            i += 1
    
    bancos.close()
    datos_indices.close()
    return 1

n_periodos = int(input("N° de periodos para analizar (min 3 - max 36): "))
print("\n")

if n_periodos >= 3 and n_periodos <= 36:
    carga_datos_sbif(n_periodos)
    print("Datos cargados")
else:
    print("Numero de periodos fuera de rango")
```

```python
# 2. Aquí se analiza la información descargada de la API, teniendo disponible 4 posibles cálculos (ver opciones desplegadas)

def leer_datos_por_banco():
    
    archivo = open("Indices_IRS.txt", "r")
    
    dato_banco = {}
    
    for linea in archivo:
        
        codigo, nombre_banco, anio, mes, irs = linea.strip().split(";")
        
        mes = "00" + mes 
        mes = mes[len(mes)-2:]
        
        if nombre_banco not in dato_banco:
            dato_banco[nombre_banco] = []
        
        dato_banco[nombre_banco].append([anio + "-" + mes, float(irs.replace(",","."))])
    
    archivo.close()
    
    return dato_banco

def leer_datos_por_periodo():
    
    archivo = open("Indices_IRS.txt", "r")
    
    dato_periodo = {}
    
    for linea in archivo:
        
        codigo, nombre_banco, anio, mes, irs = linea.strip().split(";")
        
        mes = "00" + mes 
        mes = mes[len(mes)-2:]
        
        if anio + "-" + mes not in dato_periodo:
            dato_periodo[anio + "-" + mes] = []
        
        dato_periodo[anio + "-" + mes].append(float(irs.replace(",",".")))
    
    archivo.close()
    
    return dato_periodo

def mayor_irs_banco():
    
    base_datos_banco = leer_datos_por_banco()
    
    banco_irs_mayor = ""
    periodo_irs_mayor= ""
    monto_irs_mayor = float("-inf") 
    
    for banco, dato in base_datos_banco.items():
        for per, irs in dato:
            if irs > monto_irs_mayor:
                banco_irs_mayor = banco
                periodo_irs_mayor= per
                monto_irs_mayor = irs
                
    print("El Banco", banco_irs_mayor, "presenta el mayor IRS de", monto_irs_mayor, "obtenido en el periodo del", periodo_irs_mayor)
    
def menor_irs_banco():
    
    base_datos_banco = leer_datos_por_banco()
    
    banco_irs_menor = ""
    periodo_irs_menor= ""
    monto_irs_menor = float("inf") 
    
    for banco, dato in base_datos_banco.items():
        for per, irs in dato:
            if irs < monto_irs_menor:
                banco_irs_menor = banco
                periodo_irs_menor= per
                monto_irs_menor = irs
                
    print("El Banco", banco_irs_menor, "presenta el menor IRS de", monto_irs_menor, "obtenido en el periodo del", periodo_irs_menor)

def mayor_promedio_periodo():
    
    base_datos_periodo = leer_datos_por_periodo()
    
    periodo_mayor = ""
    monto_promedio_mayor = float("-inf") 
    
    for periodo, datos in base_datos_periodo.items():
        if (sum(datos)/len(datos)) > monto_promedio_mayor:
            periodo_mayor = periodo
            monto_promedio_mayor = sum(datos)/len(datos)
    
    print("El mayor promedio del indice IRS, para los bancos chilenos, se produce en el periodo", periodo_mayor, "con un monto igual a", monto_promedio_mayor, "veces el Patrimonio Efectivo sobre Activos Ponderados por Riesgo de Crédito.")

def menor_promedio_periodo():
    
    base_datos_periodo = leer_datos_por_periodo()
    
    periodo_menor = ""
    monto_promedio_menor = float("inf") 
    
    for periodo, datos in base_datos_periodo.items():
        if (sum(datos)/len(datos)) < monto_promedio_menor:
            periodo_menor = periodo
            monto_promedio_menor = sum(datos)/len(datos)
    
    print("El menor promedio del indice IRS, para los bancos chilenos, se produce en el periodo", periodo_menor, "con un monto igual a", monto_promedio_menor, "veces el Patrimonio Efectivo sobre Activos Ponderados por Riesgo de Crédito.")

print("Seleccione el número del análisis que quiere hacer:")
print("1. Banco que muestra el índice IRS más alto durante el periodo de análisis")
print("2. Banco que muestra el índice IRS más bajo durante el periodo de análisis")
print("3. Periodo en donde el índice IRS promedio, de todos los bancos, es el más alto")
print("4. Periodo en donde el índice IRS promedio, de todos los bancos, es el más bajo")

print("\n")
alternativa_analisis = int(input("N° de la alternativa seleccionada: "))
print("\n")

if alternativa_analisis == 1:
    mayor_irs_banco()
elif alternativa_analisis == 2:
    menor_irs_banco()
elif alternativa_analisis == 3:
    mayor_promedio_periodo()
elif alternativa_analisis == 4:
    menor_promedio_periodo()
else:
    print("Debe ingresar una alternativa valida")
```
